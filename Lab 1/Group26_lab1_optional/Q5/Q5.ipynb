```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import copy
import pandas as pd
from pandas import DataFrame
import seaborn as sns 
```

```python
# a data modification function, obtain different result by changing n 
def dataprocess(old_data):
    n = 2
    new_data = copy.deepcopy(old_data)
    for k in range(len(new_data)):   
        for i in range(len(new_data[k])):
            if new_data[k][i] > n *(np.mean(old_data[k])):
                new_data[k][i] = np.median(new_data[k])
    return new_data
```

```python
num = '0';
names = ['256', '512', '1K', '2K', '4K', '8K', '16K', '32K', '64K', '128K', '256K', '512K', '1M', '2M', '4M']
```

```python
newdata00 = dataprocess(rawdata00)
newdata01 = dataprocess(rawdata01)
newdata02 = dataprocess(rawdata02)
newdata03 = dataprocess(rawdata03)

newdata10 = dataprocess(rawdata10)
newdata11 = dataprocess(rawdata11)
newdata12 = dataprocess(rawdata12)
newdata13 = dataprocess(rawdata13)
```
